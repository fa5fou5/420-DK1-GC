# Exercise 1 : Tracer une fonction

Ecrivez un programme qui trace la fonction $g(y) = e^{−y} sin(4y)$ pour $y ∈ [0, 4]$
en utilisant une ligne continue rouge. 
Utilisez 500 intervalles pour évaluer les points dans [0,4].  
Stockez toutes les coordonnées et les valeurs dans des tableaux.  
Placez le texte des graduations sur les axes et utilisez le titre "Onde sinusoïdale atténuée".

```python
# Importation
import numpy as np
import matplotlib.pyplot as plt

def g(y):
    return np.exp(-y)*np.sin(4*y)

y = np.linspace(0, 4, 501)
# définir un nouveau graphique
plt.figure()
# tracer la fonction g(y) avec ligne solide rouge
plt.plot(y, g(y), 'r-')
plt.xlabel('y'); plt.ylabel('g(y)')
plt.title(u'Onde sinusoïdale atténuée')
 # sauvgarder le grahique (format PNG et PDF)
plt.savefig("fig_ex1.png"); plt.savefig("fig_ex1.pdf")
# Afficher le graphique
plt.show()
```

# Exercise 2: Tracer deux fonctions

Comme Exercice 1, mais ajouter une courbe en pointillé noir pour la fonction $e^{-\frac{3}{2} y} \sin(4y)$.  
Inclure une légende pour chaque courbe (avec les noms $g$ et $h$).

```python
def g(y):
    return np.exp(-y)*np.sin(4*y)

def h(y):
    return np.exp(-(3./2)*y)*np.sin(4*y)

y = np.linspace(0, 4, 501)
plt.figure()
plt.plot(y, g(y), 'r-', y, h(y), 'k--')
plt.xlabel('y'); plt.ylabel('g(y)')
plt.title(u'Onde sinusoïdale atténuée')
plt.legend(['g', 'h'])

plt.savefig("fig_ex2.png"); plt.savefig("fig_ex2.pdf")
plt.show()
```
