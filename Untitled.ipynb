# Exercise 1 : Tracer une fonction
Ecrivez un programme qui trace la fonction $g(y) = e^{-y}\sin(4y)$ pour $y \in [0, 4]$ <!-- g(y) = e−ysin(4y) pour y ∈ [0, 4] -->
en utilisant une ligne continue rouge.  
Utilisez 500 intervalles pour évaluer les points dans $[0,4]$.   
Placez le texte des graduations sur les axes et utilisez le titre "Onde sinusoïdale atténuée". 
Sauvegarder la figure en une image format png intitulé "Onde_sin_att"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
y = np.linspace(0, 4, 500)
# Definir la fonction g 
def g (y):
    return np.exp(-y) * np.sin(4*y)
plt.plot(y, g(y), 'r-')
plt.title('Onde sinusoïdale atténuée')
plt.xlabel('y')
plt.ylabel('g(y)')
plt.show()
plt.savefig("Onde_sin_att.png", format = 'png')
```

# Exercise 4: Résolution d’une équation du second degré
Soit l’équation du second degré $ax2+bx+c=0$  où $a,\ b\ et\ c$ sont des coefficients réels.

* **a)** Écrivez un programme qui demande d'entrer les coefficients et affiche les solutions de l'équation.

**Indication:**
Des solutions sont recherchées dans le cas général, compte tenu du discriminant $\Delta = b^2 - 4ac$, l'équation admet comme solutions analytiques:

$\begin{cases}
\Delta > 0 : \text{ deux solutions réelles: } x_1 = \frac{-b - \sqrt{\Delta}}{2a}; x_2 = \frac{-b + \sqrt{\Delta}}{2a} \\
\Delta = 0 : \text{ une solution double: } x_0 = \frac{-b}{2a} \\
\Delta < 0 : \text{ deux solutions complexes: } z_1 = \frac{-b - i\sqrt{-\Delta}}{2a}; z_2 = \frac{-b + i\sqrt{-\Delta}}{2a}
\end{cases}$

```python
a, b, c = input("Saisire les coefficients d'un l'équation du second degré ax2 + bx +c\n").split(',')
from IPython.display import display, Math, Latex
display(Math(r' F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx'))
print(f"l'équation à résoudre est ${a}x^2+{b}x+{c}$")
```

```python
print(f"l'équation à résoudre est:")
display(Math(f"{a}x^{2}+{b}x+{c}"))
```

```python
a = float(a)
b = float(b)
c = float(c)
# Calcul de delta
delta = b**2 - 4*a*c
display(Math(f'\Delta = b^2 - 4ac = {delta}'))
```

```python
# %load solution/ex4
"""
Calcul des racines de l'equation du second degré:
a x^2 + b x + c = 0
"""
from math import sqrt
#Resultats des racines suivant la valeur de delta:
if delta > 0:
    x1 = (-b - sqrt(delta))/(2*a)
    x2 = (-b + sqrt(delta))/(2*a)
    # Affichage des solutions trouvées
    print("Les solutions sont réelles: ")
    print("La premiere racine est x1= ",x1)
    print("La seconde racines est x2= ",x2)

elif delta == 0:
    x0 = -b/(2*a)
    # Affichage de la solution trouvée
    print("Il y a une seule solution: ")
    print("La solution est", x0)

elif delta<0:
    z1 = (-b - 1j*sqrt(-delta))/(2*a)
    z2 = (-b + 1j*sqrt(-delta))/(2*a)
    # Affichage des solutions trouvées
    print("Les solutions sont complexes: ")
    print("La premiere racine est z1 = ", z1)
    print("La seconde racine est z2 = ", z2)
```
