__Question 12.__ Combien de modèle possède la marque Renault. Lister les modèles

```python
renault_models = sub_data[sub_data.Marque == 'RENAULT']
# Utiliser la méthode unique() pour obtenir la liste des modèles uniques
renault_model_list = renault_models['Modèle'].unique()
# Compter le nombre de modèles uniques
nombre_de_modeles_renault = len(renault_model_list)

# Afficher le nombre de modèles et la liste des modèles
print("Nombre de modèles possédés par la marque Renault:", nombre_de_modeles_renault)
print("Liste des modèles Renault:", renault_model_list)
```

**Question 15.** Tracer un bar chart dont l'axe des X représente le Modèle. Intitullé 'Nombre de voitures Renault testés selon le modèle'.  
Ajouter:
1. label pour l'axe des y: 'Count'
2. label pour l'axe des x: 'Modèles Renault' avec un angle de 45 deg
3. Liste des nom de modèle dans l'axe des x
4. Couleur rouge des bar
5. Contour de couleur noir pour les barres
**Résultat**



```python
import matplotlib.pyplot as plt
```

```python
plt.bar(renault_cars.Modèle, renault_cars.Count, color='red', edgecolor='black')
plt.xlabel('Modèle Renault')
plt.ylabel('Count')
plt.xticks(range(len(renault_cars)), renault_cars['Modèle'], rotation=45)
plt.title('Nombre de voitures Renault testés selon le modèle')
plt.get_cmap('viridis')
# Add the count on top of each bar
for i, bar in enumerate(renault_cars):
    plt.text(i, bar.get_height() + .5, grouped_counts['count'].iloc[i], ha='center')

plt.show()
```
