**Question 13.** Lister le nombre des voitures par groupement selon la marque et le modèle. Enregistrer le résultat dans une variable appelée `grouped_counts`

**Astuce:** Utiliser le lien [Comparaison avec sql](https://pandas.pydata.org/pandas-docs/stable/getting_started/comparison/comparison_with_sql.html) du site pandas pour determiner la solution.
