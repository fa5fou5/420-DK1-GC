```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('Car_sales.csv') as f:
    print(f.read())
```

```python
with open('Car_sales.csv') as f:
    next(f)
    manufacturer = []
    date = []
    sales = []
    for line in f.readlines():
        manufacturer.append(line.strip().split(';')[0])
        date.append(line.strip().split(';')[1])
        sales.append(line.strip().split(';')[2])
```

```python
print(manufacturer, type(manufacturer))
print(date, type(date))
print(sales, type(sales))
```

```python
manufacturer = np.array(manufacturer)
date = np.array(date, dtype = 'datetime64')
sales = np.array(sales, dtype = float)
```
