```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label=r'$\sin(x)$')  # Utilisation de balises LaTeX pour la lgende
plt.plot(x, y2, label=r'$\cos(x)$')  # Utilisation de balises LaTeX pour la lgende
plt.xlabel(r'$x$')  # Utilisation de balises LaTeX pour l'tiquette de l'axe x
plt.ylabel(r'$y$')  # Utilisation de balises LaTeX pour l'tiquette de l'axe y
plt.title(r'Fonctions trigonomtriques')  # Utilisation de balises LaTeX pour le titre
plt.legend()  # Affichage de la lgende
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')  # Lgende pour sin(x)
plt.plot(x, y2, label='cos(x)')  # Lgende pour cos(x)
plt.xlabel('x')  # tiquette de l'axe x
plt.ylabel('y')  # tiquette de l'axe y
plt.title('Fonctions trigonomtriques')  # Titre
plt.legend()  # Affichage de la lgende
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)
y = np.exp(-x**2)

plt.plot(x, y, label=r'$e^{-x^2}$')  # Utilisation de balises LaTeX pour la lgende
plt.xlabel(r'$x$')  # Utilisation de balises LaTeX pour l'tiquette de l'axe x
plt.ylabel(r'$y$')  # Utilisation de balises LaTeX pour l'tiquette de l'axe y
plt.title(r'Fonction Gaussienne')  # Utilisation de balises LaTeX pour le titre
plt.legend()  # Affichage de la lgende
plt.show()
```
