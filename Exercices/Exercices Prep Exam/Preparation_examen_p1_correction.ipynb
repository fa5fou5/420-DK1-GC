__Question 3:__ Que represente le type de données `object`? Quels sont les type de données de base en Pandas?

En Pandas, le type `object` est généralement utilisé pour stocker des données de type texte (chaînes de caractères), mais il peut aussi contenir d’autres types de données Python comme des listes ou des dictionnaires. Lorsque vous voyez un dtype `object` dans un DataFrame Pandas, cela signifie que la colonne contient des données qui sont traitées comme des objets Python génériques.

Voici quelques points clés à retenir sur le type object en Pandas :

`Polyvalence :` Une colonne de type object peut mélanger des données textuelles et non textuelles.  
`Compatibilité :` Le type object assure une compatibilité avec d’autres bibliothèques Python.  
`Efficacité :` Bien que le type object soit plus flexible, il peut être moins efficace en termes de stockage et de vitesse par rapport aux types de données spécifiques comme `int` ou `string`.

[Type de données texte en Pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html#text-types)  
[Type de données basique en Pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/basics.html#basics-dtypes)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'quartier': ['Q1', 'Q2', 'Q3', 'Q4'],
    'temperature_max': [29.27, 26.45, 27.11, 26.92],
    'Kwh': [32.88, 25.19, 24.36, 26.77]
}
```
