# Exercise 1 : Tracer une fonction

* __a.__ Ecrivez un programme qui trace la fonction $g(y) = e^{−y} sin(4y)$ pour $y ∈ [0, 4]$
en utilisant une ligne continue rouge.
* __b.__ Utilisez 500 intervalles pour évaluer les points dans [0,4].  
* __c.__ Placez le texte des graduations sur les axes et utilisez le titre "Onde sinusoïdale atténuée".
* __d.__ Sauvegarder le graphique en format png. 

```python
import numpy as np
from numpy import sin, exp
import matplotlib.pyplot as plt
```

```python
# definition de l'intervalle
y = np.linspace(0,4, 501)
```

```python
# definition de la fonction g
def g(v):
    return exp(-v)*sin(4*v)
```

```python
#g = exp(-y)*sin(4*y)
```

```python
type(g(y))
```

```python
x= np.linspace(0,10, 101)
```

```python
%matplotlib inline
```

```python
plt.plot(y, sin(4*y),'k--')
plt.plot(y, g(y),'r-')
plt.title('Onde sinusoïdale atténuée')
plt.xlabel('y')
plt.ylabel('g(y)')
plt.grid()
```

```python
# Sauvegarde de la figure
plt.savefig('Onde sinusoïdale atténuée.png', format= 'png')
```

```python
plt.plot(x, g(x),'m--')
```
