# Scatter plot

```python
# génération de données
N = 50
x = np.linspace(0., 10., N)
y = np.sin(x)**2 + np.cos(x)
```

```python
# plt.figure()
plt.scatter(x, y, s=15, label = r'$y = sin(x^2) +cos(x)$', color = 'r' )
plt.axis('equal')
plt.legend()
plt.xlabel(r'$x (rad)$')
plt.ylabel(r'$y$')
```

```python
plt.savefig('scatter2.png', dpi = 300, bbox_inches = 'tight', facecolor='w')
```

```python
np.random.seed(100)
N = 50
randx = np.random.random(N) * 100
randy = np.random.random(N) * 100
plt.figure(figsize=(9, 6))
plt.scatter(randx, randy, marker=r'$\alpha$', s = 50, color = 'darkorange')
plt.axis('equal')
plt.xlabel('randx')
plt.ylabel('randy')
plt.tight_layout()
```
