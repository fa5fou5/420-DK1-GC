```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# lecture du fichier sales

with open('company_sales_data.csv', newline = '') as f:
    next(f) # sauter 1er ligne
    month_number = []
    total_profit = []
    
    for line in f.readlines():
        row = line.strip().split(',')
        # print(line)
        # print(row)
        month_number.append(row[0])
        total_profit.append(row[-1])

month_number = np.array(month_number, dtype = int)
total_profit = np.array(total_profit, dtype = int)
print(month_number, type(month_number))
print(total_profit, type(total_profit))
    
```

```python
plt.plot(month_number, total_profit)
```
