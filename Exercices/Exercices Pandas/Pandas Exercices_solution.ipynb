## 6. Ressortir les élèments de la série A qui ne sont pas présents dans la série B ?

```python
# En soustrayant les élèment de la ser2 de la ser1
# Input
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
# Solution
ser1[~ser1.isin(ser2)]
```

```python
ser1.isin(ser2)
```

__Note:__ L'opérateur unaire `~` n'est pas spécifique à Pandas ; il s'agit d'un opérateur Python standard qui est utilisé pour le complément logique. Cela signifie que vous pouvez l'utiliser avec des types de données booléens, tels que des tableaux (listes, tableaux NumPy, etc.) et des séries (comme les séries Pandas et les tableaux de booléens NumPy).
Voici quelques exemples pour illustrer que l'opérateur `~` peut être utilisé en dehors de Pandas :

1. Avec les listes en Python :
>~~~Python
>boolean_list = [True, False, False, True]
>complement_list = [~b for b in boolean_list]
>print(complement_list)  # Résultat : [False, True, True, False]
>~~~
2. Avec les tableaux NumPy :
>~~~Python
>import numpy as np
>boolean_array = np.array([True, False, False, True])
>complement_array = ~boolean_array
>print(complement_array)  # Résultat : [False  True  True  False]
>~~~
3. Avec d'autres types de données booléens en Python :
>~~~Python
>boolean_value = True
>complement_value = ~boolean_value
>print(complement_value)  # Résultat : False
>~~~

Comme vous pouvez le voir, l'opérateur `~` peut être appliqué à différents types de données booléens en dehors de Pandas, étant donné que c'est un opérateur logique standard en Python. Cela donne des résultats similaires à ceux obtenus avec Pandas.

## Comment obtenir les articles qui ne sont pas communs aux séries A et B ?

```python
# Input
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
# Solution
ser_u = pd.Series(np.union1d(ser1, ser2))  # union
ser_i = pd.Series(np.intersect1d(ser1, ser2))  # intersect
ser_u[~ser_u.isin(ser_i)]
```
